## Fetch and process multiple readings

```python
from spectral_board_manager.spectral_sensor import SpectralSensor
from spectral_board_manager.data_parser import SpectralAnalysis

PORT1 = "/dev/cu.wchusbserial10"
BOARD_ID = "board-2"
SENSORS = 10 # max 16
board1 = SpectralSensor(PORT1)
analyser = SpectralAnalysis()

for i in range(1, SENSORS+1):
    data = board1.read_sensor(i)
    analyser.parse_new_data(data, "board-2")
    rgb, hex = analyser.plot_normalised_spectrum()
    analyser.append_to_csv(hex)
```

## Tune sensor settings 

Options:
- **LEDS_ON**: For measuring colour of solid / reflective surfaces. If True, Onboard LEDs stay on during measurments. Defaults to False, to flash once before measurement.

- **GAIN**: Can be set to 1, 2, 4 or 8. Sets the analogue amplification applied to the photodiode signals before ADC conversion; higher gain increases sensitivity to low light but saturates earlier and amplifies noise.

- **ATIME (integration time)**: Defines the base duration over which photons are accumulated by each photodiode; longer ATIME increases signal level and SNR but reduces maximum measurable intensity.

- **ASTEP (integration step size)**: Sets the number of clock cycles per ATIME step, allowing fine-grained scaling of the effective integration time without changing gain.

**Effective integration time = (ATIME + 1) × (ASTEP + 1) × 2.78 µs**

```python
LEDS_ON = True
GAIN = 2
ATIME = 30 # unitless: 0–255
ASTEP = 600 # unitless: 0–65535

board1.set_leds_on_during_measurements(LEDS_ON)
board1.set_sensor_settings(GAIN, ATIME, ASTEP)
```
